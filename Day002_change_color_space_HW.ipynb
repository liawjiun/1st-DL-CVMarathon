```python
import numpy as np
import cv2

img_file = 'Part01/GF_4.jpg'

# 以彩色圖片的方式載入
img = cv2.imread(img_file, cv2.IMREAD_COLOR)
img.shape
```

```python
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_Lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img_hls)
```

```python
plt.imshow(img_Lab)
```
