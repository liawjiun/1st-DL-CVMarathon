### 嘗試用 keras 的 DepthwiseConv2D 等 layers 實做 Inverted Residual Block.
   - depthwise's filter shape 爲 (3,3), padding = same
   - 不需要給 alpha, depth multiplier 參數
   - expansion 因子爲 6

##### 載入套件

```python
from keras.models import Input, Model
from keras.layers import DepthwiseConv2D, Conv2D, BatchNormalization, ReLU, Add
```

##### 定義 Separable Convolution 函數 (請在此實做)

```python
input = Input((64, 64, 3))
int(input.shape[3])
```

```python
def InvertedRes(input, expansion):
  x = Conv2D(filters=expansion,kernel_size=1, padding = 'same', activation=None)(input)
  x = DepthwiseConv2D(kernel_size=3, padding = 'same', activation=None)(x)
  x = BatchNormalization() (x) 
  x = ReLU(max_value=None, negative_slope=0.0, threshold=0.0)(x)
  x = Conv2D(filters=int(input.shape[3]),kernel_size=1, padding = 'same', activation=None)(x)
  x = BatchNormalization() (x) 
  x = ReLU(max_value=None, negative_slope=0.0, threshold=0.0)(x)
  x = Add()([x, input])

  return x
```

##### 建構模型

```python
input = Input((64, 64, 3))
output = InvertedRes(input, 6)
model = Model(inputs=input, outputs=output)
model.summary()
```
