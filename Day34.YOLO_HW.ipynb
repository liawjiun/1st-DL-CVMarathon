## Day34.YOLO 細節理解 - 損失函數
今天的課程，我們講述了
* 損失函數是描述模型預測出來的結果和實際的差異的依據    
* YOLO 損失函數的設計包含物件位置的定位與物件類別辨識    
* YOLO損失函數透過超參數設定模型有不同的辨識能力    



### 作業
仔細觀察，bbox 寬高計算損失方式和bbox中心計算損失方式有哪邊不一樣嗎? 為什麼要有不同的設計?
![title](loss function.png)



Ans: 
(1) 寬高計算: 先各自取根號, 再相減, 再平方.
(2) 中心計算: 直接相減, 再取平方.
(3) 為何需要不同設計:  大物體和小物體, 中心預測如果都偏差2-pixel, 就是偏差了2-pixel.  但是對於寬和高, 如果預測都誤差5-pixel, 對於小物體(例如WxH = 10x10), 就相當於誤差50%. 但是對於大物體(例如(WxH = 100x100), 則只相當於誤差5%.  所以誤差的比重不同. 若是直接相減, 無法反映出同是誤差5-pixel, 對小物體框的誤差較大. => 作者利用開根號, 來降低對大物體的權重. (sqrt(15)- sqrt(10) = 0.71(小物體寬高誤差5-pixel的比重);   而sqrt(105) - sqrt(100) = 0.24(小物體寬高誤差5-pixel的比重)
